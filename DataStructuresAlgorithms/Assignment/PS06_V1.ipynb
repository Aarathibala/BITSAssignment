```python
class AdmissionCounters(object):
    #To initialize the object
    def __init__(self,c,n):
        print('init')
        self.c = c
        self.n = n
        self.queues = [[None for i in range(n)] for  j in range(c)]
        self.starts = [0 for i in range(c)]
        self.ends = [0 for i in range(c)]
        self.sizes = [0 for i in range(c)]
        self.open = [False for i in range(c)]
        self.open[0] = True
    
    def isOpen (self, counterId):
        print('isOpen' + str(counterId))
        return self.open[counterId-1]
    
    def getCounter(self, counterID):
        print('getCounter' + str(counterID))
        return self.queues[counterID - 1]
    
    def addStudent(self, studentID):
        print('add Student'+ str(studentID))
        fullcounter = True
        min=0
        for i in range(self.c):
            if(self.sizes[i]) != self.n:
                fullcounter = False 
                break
        if fullcounter:
            return -1
        for i in range(self.c):
            if(self.open[i] and self.sizes[i] < self.sizes[min]):
                min = i
        if(self.sizes[min]==self.n):
            min = self.openNextCounter()
        print('min' + str(min))
        if(min != -1):
            self.queueEnqueue(min,studentID)
        return min;

    def openNextCounter(self):
        fullCounter = True
        for i in range(self.c):
            if not self.open[i]:
                fullCounter = False
                return i;
        if fullCounter:
            return -1
    def giveAdmissionTicket (self):
        numberOfTicketsIssued = 0;
        for i in range(self.c):
            if self.open[i]:
                numberOfTicketsIssued += self.queueDequeue(i)
        return numberOfTicketsIssued
            
    def queueEnqueue(self, counterId, studentID):
        print('queueEnqueue' + str(counterId) + ' ' + str(studentID))
        # Check queue is full or not
        if(self.sizes[counterId] == self.n):
            print('\n'+ counterId +'This Counter is full')
 
        # Insert element at the rear
        else:
            #print(self.queues[counterId])
            self.open[counterId] = True
            self.queues[counterId][self.ends[counterId]] = studentID
            #print(self.queues[counterId])
            self.ends[counterId] = (self.ends[counterId] + 1) % self.n
            self.sizes[counterId] += 1
            #self.tail = (self.tail + 1) % self.maxSize
            #self.queue.append(data)
            #self.rear += 1
    def queueDequeue(self, counterId):
        isDequeued = 0         
        # If queue is empty
        if(self.sizes[counterId] == 0):
            print("Queue is empty")
            isDequeued = 0
 
        # Pop the front element from list
        else:
            self.queues[counterId][self.starts[counterId]] = None
            self.starts[counterId] = (self.starts[counterId] + 1) % self.n
            self.sizes[counterId] -= 1
            isDequeued = 1
            #x = self.queue.pop(0)
            #self.rear -= 1
        return isDequeued
            
# Using readlines()
inputFile = open('inputPS6.txt', 'r')
Lines = inputFile.readlines()
count = 0
# writing to file
outputFile = open('outputPS6.txt', 'w')
inputKeys = []
inputKeysLength = 0
print('Before For Loop')
# Strips the newline character
for line in Lines:
    if not line.strip():
        print('Empty Line')
    else:
        print('Non Empty Line')
        print(line.strip())
        if(count == 0):
            # check if 0 index is ticketSystem
            maxNumberOfCounters=line.strip().split(':')[1]
            maxNumberOfStudentsPerCounter=line.strip().split(':')[2]
            admissionCounters = AdmissionCounters(int(maxNumberOfCounters),int(maxNumberOfStudentsPerCounter))
            count += 1
        else:
            inputKeys = line.strip().split(':')
            inputKeysLength = len(inputKeys)
            if(inputKeysLength > 1 and inputKeys[0] == 'isOpen'):
                outputFile.write(line.strip()+ ' >> ' + str(admissionCounters.isOpen(int(inputKeys[1]))) +'\n')
                count += 1
            elif(inputKeysLength > 1 and inputKeys[0] == 'getCounter'):
                outputFile.write(line.strip()+ ' >> ' + str(admissionCounters.getCounter(int(inputKeys[1]))) +'\n')
                count += 1
            elif(inputKeysLength > 1 and inputKeys[0] == 'addStudent'):
                addStudent = admissionCounters.addStudent(int(inputKeys[1]))
                if addStudent == -1:
                    outputFile.write(line.strip()+ ' >> ' + 'all queues are full' +'\n')
                    count += 1
                else:
                    outputFile.write(line.strip()+ ' >> c' + str(addStudent+1) +'\n')
                    count += 1
            elif(inputKeys[0] == 'giveAdmissionTicket'):
                outputFile.write(line.strip()+ ' >> ' + str(admissionCounters.giveAdmissionTicket()) +'\n')
                count += 1
        
    #print("Line{}: {}".format(count, line.strip()))
    #outputFile.write(line.strip()+'\n')
    

print("Max values are {} and {}".format(maxNumberOfCounters, maxNumberOfStudentsPerCounter))
#print(admissionCounters.open[0])
print(admissionCounters.queues)

outputFile.close()

```
